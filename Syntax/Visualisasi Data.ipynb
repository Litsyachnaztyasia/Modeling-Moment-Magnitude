# Load Package and Data

```python
import pandas as pd

import plotly.express as px
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('../Data/M7 Declustering.csv')
data2 = pd.read_csv('../Data/M8 Declustering.csv')
data3 = pd.read_csv('../Data/M9 Declustering.csv')
```

# Function

Descriptive function

```python
def desc_stat(df):
    # Menampilkan waktu terkecil dan terbesar
    min_time = df['time'].min()
    max_time = df['time'].max()
    
    print(f"Waktu terkecil: {min_time}")
    print(f"Waktu terbesar: {max_time}")
    
    # Menghitung statistik deskriptif
    numerical_columns = df[['latitude', 'longitude', 'depth', 'mag']]
    descriptive_stats = numerical_columns.describe()
    
    print("\nDescriptive Stats:")
    print(descriptive_stats)
    
    print("\nBanyaknya kolom dan baris:")
    print(df.shape)
```

Plot function

```python
def plot_summary(df, kolom, lat_col='latitude', lon_col='longitude', date_col='date', title_prefix=''):
    # Plot Time Series (scatter, tanpa garis)
    if date_col in df.columns:
        waktu = pd.to_datetime(df[date_col], errors='coerce')
        plt.figure(figsize=(7,4))
        plt.scatter(waktu, df[kolom], color='tab:blue')
        plt.xlabel('Date')
        plt.ylabel(kolom)
        plt.title(f'{title_prefix}Time Series {kolom}')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    
    # Plot Histogram
    plt.figure(figsize=(7,4))
    plt.hist(df[kolom], bins=30, color='orange', edgecolor='k')
    plt.xlabel(kolom)
    plt.ylabel('Frekuensi')
    plt.title(f'{title_prefix}Histogram {kolom}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Scatter Mapbox
    if lat_col in df.columns and lon_col in df.columns:
        fig = px.scatter_mapbox(
            df,
            lat=lat_col,
            lon=lon_col,
            color=kolom,
            color_continuous_scale="Inferno",
            zoom=5,
            height=500,
            width=800,
            title=f'{title_prefix}Scatter Mapbox {kolom}',
            mapbox_style="open-street-map"
        )
        fig.show()
```

# M7

Descriptive statistic

```python
desc_stat(data1)
```

## Magnitude

```python
plot_summary(data1, 'mag')
```

## Depth

```python
plot_summary(data1, 'depth')
```
