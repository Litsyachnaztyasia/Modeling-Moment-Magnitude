# Load Package and Data

```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('../Data/Rekap Hasil Simulasi.xlsx', sheet_name='olah data')

# Display the first few rows
df.head()
```

```python
import matplotlib.pyplot as plt

# Grafik 1: M7 Segment Simulation
plt.figure(figsize=(10, 6))
plt.plot(df['n'], df['logn_m7'], label='Lognormal-GPD', color='red')
plt.plot(df['n'], df['norm_m7'], label='Normal-GPD', color='blue')
plt.hlines(0.749, df['n'].min(), df['n'].max(), colors='black', linestyles='dotted', label='Historical b-value')
plt.hlines(1.150, df['n'].min(), df['n'].max(), colors='black', linestyles='dashdot', label='PUSGEN b-value')
plt.title('Average b-value for M7 Segment Simulation')
plt.xlabel('Number of Simulation')
plt.ylabel('Average b-value')
plt.ylim(0, 1.2)
plt.legend()
plt.grid(True)
plt.show()

# Grafik 2: M8 Segment Simulation
plt.figure(figsize=(10, 6))
plt.plot(df['n'], df['burr_m8'], label='Burr-GPD', color='red')
plt.plot(df['n'], df['loglog_m8'], label='Loglogistic-GPD', color='blue')
plt.plot(df['n'], df['logn_m8'], label='Lognormal-GPD', color='green')
plt.hlines(0.667, df['n'].min(), df['n'].max(), colors='black', linestyles='dotted', label='Historical b-value')
plt.hlines(1.080, df['n'].min(), df['n'].max(), colors='black', linestyles='dashdot', label='PUSGEN b-value')
plt.title('Average b-value for M8 Segment Simulation')
plt.xlabel('Number of Simulation')
plt.ylabel('Average b-value')
plt.ylim(0, 1.2)
plt.legend()
plt.grid(True)
plt.show()

# Grafik 3: M9 Segment Simulation
plt.figure(figsize=(10, 6))
plt.plot(df['n'], df['loglog_m9'], label='Loglogistic-GPD', color='red')
plt.plot(df['n'], df['logn_m9'], label='Lognormal-GPD', color='blue')
plt.plot(df['n'], df['norm_m9'], label='Normal-GPD', color='green')
plt.hlines(0.751, df['n'].min(), df['n'].max(), colors='black', linestyles='dotted', label='Historical b-value')
plt.hlines(1.080, df['n'].min(), df['n'].max(), colors='black', linestyles='dashdot', label='PUSGEN b-value')
plt.title('Average b-value for M9 Segment Simulation')
plt.xlabel('Number of Simulation')
plt.ylabel('Average b-value')
plt.ylim(0, 1.2)
plt.legend()
plt.grid(True)
plt.show()
```
